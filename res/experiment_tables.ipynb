```python
import pandas as pd

print("The result for normal mode...")
seahorn_norm = pd.read_csv('seahorn.csv')
mean_time = seahorn_norm['Timing (Seconds)'].mean()
tol_time = seahorn_norm['Timing (Seconds)'].sum()
count_passed = 0
for j in range(seahorn_norm.shape[0]):
    if seahorn_norm['Result'][j] == "passed":
        count_passed += 1
pass_str = "{pas}/{total}".format(pas=count_passed,total=seahorn_norm.shape[0])
print(f'Total time: {tol_time:.2f}, Mean time: {mean_time:.2f}, Passed/Total: {pass_str}')
seahorn_norm
```
